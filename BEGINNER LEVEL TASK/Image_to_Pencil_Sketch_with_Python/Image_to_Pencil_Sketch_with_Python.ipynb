```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('/content/dog.jpg')

# Convert the image to grayscale
b_and_w_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted_image = 255 - b_and_w_image

# Blur the inverted image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

# Invert the blurred image
inverted_blurred = 255 - blurred

# Display the final result using Matplotlib
plt.imshow(inverted_blurred, cmap='gray')
plt.axis('off')  # Hide the axis
plt.title('Inverted Blurred Image')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('/content/dog.jpg')

# Convert the image to grayscale
b_and_w_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted_image = 255 - b_and_w_image

# Blur the inverted image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

# Invert the blurred image
inverted_blurred = 255 - blurred

# Create the pencil sketch
pencil_sketch = cv2.divide(b_and_w_image, inverted_blurred, scale=256.0)

# Display the pencil sketch using Matplotlib
plt.imshow(pencil_sketch, cmap='gray')
plt.axis('off')  # Hide the axis
plt.title('Pencil Sketch')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

# Path to your image file
image_path = '/content/dog.jpg'

# Load the image
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image. Please check the file path.")
else:
    # Convert the image to grayscale
    b_and_w_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Invert the grayscale image
    inverted_image = 255 - b_and_w_image

    # Blur the inverted image
    blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

    # Invert the blurred image
    inverted_blurred = 255 - blurred

    # Create the pencil sketch
    pencil_sketch = cv2.divide(b_and_w_image, inverted_blurred, scale=256.0)

    # Display the pencil sketch using Matplotlib
    plt.imshow(pencil_sketch, cmap='gray')
    plt.axis('off')  # Hide the axis
    plt.title('Pencil Sketch')
    plt.show()
```
